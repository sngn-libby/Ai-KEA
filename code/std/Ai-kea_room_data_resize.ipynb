```
from PIL import Image

image = pathlib.Path('/content/drive/Shared drives/scsa_2019_b/Project_Ai-KEA/data/room/livingroom')
image = list(image.glob('*.jpg'))
for i in range(len(image)):
    im = Image.open(str(image[i])).convert('RGB')
    resize_image = im.resize((im.size[0]//2,im.size[1]//2))
    resize_image.save('/content/drive/Shared drives/scsa_2019_b/Project_Ai-KEA/data_resize/room/{0}/{1}{2:03d}.jpg'.format('livingroom','livingroom',i))
    print('livingroom', i)
```
